```
import pandas as pd 


data=pd.read_csv('/content/drive/My Drive/Deep Learning /airquality.csv')
print(data.shape)

data=pd.melt(frame=data,id_vars=['Month','Day'],var_name="measurement",value_name="readings")#for turning columns into rows
# print(data.head(50))
print(data.shape)
#melting and pivoting are two different methods for analysis of data 
#pivot turns unique rows data in to columns for better analysis 

pivot_table=pd.pivot_table(data,index=['Month','Day'],columns='measurement',values='readings')
# print(pivot_table.head(10))
print(pivot_table.shape)

#for reseting data again 
pivot_table=pivot_table.reset_index()
# print(pivot_table)

```
