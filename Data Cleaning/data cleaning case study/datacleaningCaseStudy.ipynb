```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

#read data through pandas 
data=pd.read_csv('/content/drive/My Drive/Deep Learning /gapminder.csv')



life=data['Life expectancy']

#separate data according to era's 
gapminder_1800=data.iloc[:,1:101]
gapminder_1900=data.iloc[:,101:201]
gapminder_2000=data.iloc[:,201:]

gapminder_1800['Life expectancy']=life
gapminder_1900['Life expectancy']=life

#drop missing values 
gapminder_1800=gapminder_1800.dropna(how='any')
gapminder_1900=gapminder_1900.dropna(how='any')
gapminder_2000=gapminder_2000.dropna(how='any')


#joined again 
gap_minder=pd.concat([gapminder_1800,gapminder_1900,gapminder_2000])



#change columns in to rows for better reading 
gap_minder_melt=pd.melt(gap_minder,id_vars='Life expectancy')

#name of columns after melting 
gap_minder_melt.columns=["Country",'Years','Life expectancy']

print(gap_minder_melt)

#visualize the data 
plt.subplot(2,1,1)

#frequency of life 
gap_minder_melt['Life expectancy'].plot(kind='hist')


#group by the years and take their mean 
gapminder_agg=gap_minder_melt.groupby('Years')['Life expectancy'].mean()
print(gapminder_agg)
plt.subplot(2,1,2)
gapminder_agg.plot()
plt.title('Life expectancy over the years')
plt.ylabel('Life expectancy')
plt.xlabel('Year')

# Display the plots
plt.tight_layout()

plt.show()







```
